```python
from google.colab import files
import fitz  # pymupdf

print("üìÑ Upload your resume (PDF format)")
uploaded_resume = files.upload()
resume_path = list(uploaded_resume.keys())[0]

def extract_text_from_pdf(path):
    doc = fitz.open(path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

resume_text = extract_text_from_pdf(resume_path)
print("Resume text extracted successfully!")
```

```python
from openai import OpenAI
# Replace with own API key
api_key =
client = OpenAI(api_key=api_key)
```

```python
import re

def extract_score(response_text):
    match = re.search(r"(\d{1,3})", response_text)
    return int(match.group(1)) if match else 0

def match_jobs(resume, job_descriptions):
    results = []
    for job in job_descriptions:
        prompt = f"""Given the resume:
---
{resume}

Evaluate its fit for the job below and rate from 0-100. Explain briefly.
---
{job['description']}
"""
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[{"role": "user", "content": prompt}]
        )
        content = response.choices[0].message.content
        score = extract_score(content)
        results.append({**job, "score": score, "reason": content})
    return sorted(results, key=lambda x: x['score'], reverse=True)

def generate_cover_letter(resume, job_description):
    prompt = f"""Write a personalized, enthusiastic cover letter based on this resume:
---
{resume}

For the following job:
---
{job_description}
"""
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content
```

```python
job_descriptions = [
    {
        "title": "Backend Engineer",
        "company": "TechNova",
        "description": "We‚Äôre looking for someone skilled in Python, Flask, and PostgreSQL to help build our API layer."
    },
    {
        "title": "Data Scientist",
        "company": "Insightful AI",
        "description": "Seeking a data scientist with machine learning, Pandas, and visualization experience. Bonus for NLP skills."
    }
]
```

```python
matches = match_jobs(resume_text, job_descriptions)

for job in matches:
    print(f"\nüìå Job: {job['title']} at {job['company']}")
    print(f"üî¢ Match Score: {job['score']}")
    print(f"üó£ Reason: {job['reason']}")
    print("\nüìù Generating cover letter...\n")
    print(generate_cover_letter(resume_text, job['description']))
    print("\n" + "="*80 + "\n")
```
